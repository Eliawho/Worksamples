```
# create a function that retrieves the news stories, headline, author ,synopsis
# and data and time for each story

#fetch the code
import requests
url = 'https://www.space.com/news'
response = requests.get(url)
if(response.ok):
  data = response.text

from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')

for i in range(1, 22):
  pass
  tags = 'listingResult small result' + str(i) + " "
  article = soup.find(class_ = tags)
  if article is not None:
    title = article.find('h3').get_text()
    print(f"Title: {title}")
    author = article.select_one('.by-author > span').get_text()
    print(f"Author: {author}")
    publish_time = article.select_one('.byline > time')["datetime"]
    print(f"Publish Time: {publish_time}")
    synopsis = article.select_one('.synopsis').get_text()
    print(f"Synopsis: {synopsis}")

```
